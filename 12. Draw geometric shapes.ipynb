# Write a program in python to draw different geometirc shapes and to wirte text on images using OpenCV

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
img = np.zeros((400, 400, 3), dtype = "uint8")
# plt.imshow(img) # cv2. imshow('draw shapes', img)
# plt.show() # cv2.waitKey(0)
cv2.line(img, (10, 100), (100, 100), (0, 0, 255), 5)

plt.imshow(img)
plt.show()
```

```python
img2 = np.zeros((400, 400, 3), dtype="uint8")
cv2.rectangle(img2, (30, 30), (300, 200), (0, 255, 0), 5)
plt.imshow(img2)
plt.show()
```

```python
img3 = np.zeros((400, 400, 3), dtype="uint8")
cv2.circle(img3, (200, 200), 80, (255, 0, 0), 3)
plt.imshow(img3)
plt.show()
```

```python
img4 = np.zeros((400, 400, 3), dtype="uint8")
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img4, 'Meow', (50, 50),
           font, 0.8, (0, 250, 250), 2, cv2.LINE_AA)
plt.imshow(img4)
plt.show()
```
